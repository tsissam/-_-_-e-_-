# Données pour créer les graphiques

CRS (Center of School Resources):
Boys, Level 2: 18.5% have a math score >= 80%, 55.6% can read a paragraph.
Girls, Level 2: 34.8% have a math score >= 80%, 50.0% can read a paragraph.
Total, Level 2: 28.8% have a math score >= 80%, 52.1% can read a paragraph.
CRS (Center of School Resources):
Boys, Level 3: 65.6% have a math score >= 80%, 78.1% can read a paragraph.
Girls, Level 3: 54.8% have a math score >= 80%, 66.1% can read a paragraph.
Total, Level 3: 58.5% have a math score >= 80%, 70.2% can read a paragraph.
EP (Elementary School):
Boys, Level 3: 23.7% have a math score >= 80%, 35.3% can read a paragraph.
Girls, Level 3: 22.7% have a math score >= 80%, 28.1% can read a paragraph.
Total, Level 3: 23.2% have a math score >= 80%, 31.6% can read a paragraph.
EP (Elementary School):
Boys, Level 5: 60.7% have a math score >= 80%, 75.2% can read a paragraph.
Girls, Level 5: 52.6% have a math score >= 80%, 63.4% can read a paragraph.
Total, Level 5: 56.4% have a math score >= 80%, 68.9% can read a paragraph.

```python
import numpy as np
import matplotlib.pyplot as plt

# Données
categories_CRS = ['Boys, Level 2', 'Girls, Level 2', 'Total, Level 2',
                  'Boys, Level 3', 'Girls, Level 3', 'Total, Level 3']

categories_EP = ['Boys, Level 3', 'Girls, Level 3', 'Total, Level 3',
                 'Boys, Level 5', 'Girls, Level 5', 'Total, Level 5']

math_scores_CRS = [18.5, 34.8, 28.8, 65.6, 54.8, 58.5]
reading_scores_CRS = [55.6, 50.0, 52.1, 78.1, 66.1, 70.2]

math_scores_EP = [23.7, 22.7, 23.2, 60.7, 52.6, 56.4]
reading_scores_EP = [35.3, 28.1, 31.6, 75.2, 63.4, 68.9]

x_CRS = np.arange(len(categories_CRS))
x_EP = np.arange(len(categories_EP))
bar_width = 0.35

# Créer les graphiques
plt.figure(figsize=(14, 6))

# Graphique CRS
plt.subplot(1, 2, 1)
plt.bar(x_CRS - bar_width/2, math_scores_CRS, bar_width, label='Math Scores', color='skyblue')
plt.bar(x_CRS + bar_width/2, reading_scores_CRS, bar_width, label='Reading Scores', color='salmon')

# Ajouter les labels des pourcentages
for i in range(len(categories_CRS)):
    plt.text(x_CRS[i] - bar_width/2, math_scores_CRS[i] + 1, f"{math_scores_CRS[i]}%", ha='center', va='bottom')
    plt.text(x_CRS[i] + bar_width/2, reading_scores_CRS[i] + 1, f"{reading_scores_CRS[i]}%", ha='center', va='bottom')

plt.ylabel('Scores (%)')
plt.title('CRS Performance by Gender and Level')
plt.xticks(x_CRS, categories_CRS, rotation=45, ha='right')
plt.legend(loc='upper left', bbox_to_anchor=(0,1))

# Graphique EP
plt.subplot(1, 2, 2)
plt.bar(x_EP - bar_width/2, math_scores_EP, bar_width, label='Math Scores', color='skyblue')
plt.bar(x_EP + bar_width/2, reading_scores_EP, bar_width, label='Reading Scores', color='salmon')

# Ajouter les labels des pourcentages
for i in range(len(categories_EP)):
    plt.text(x_EP[i] - bar_width/2, math_scores_EP[i] + 1, f"{math_scores_EP[i]}%", ha='center', va='bottom')
    plt.text(x_EP[i] + bar_width/2, reading_scores_EP[i] + 1, f"{reading_scores_EP[i]}%", ha='center', va='bottom')

plt.ylabel('Scores (%)')
plt.title('EP Performance by Gender and Level')
plt.xticks(x_EP, categories_EP, rotation=45, ha='right')
plt.legend(loc='upper left', bbox_to_anchor=(0,1))

plt.tight_layout()
plt.show()
```
